```python
#import ML libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression  
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier


#for model assessment
from sklearn.metrics import confusion_matrix, classification_report

#for vizualization
from matplotlib.pyplot import figure
```

```python
#prepare independent and dependent variables
df=pd.read_csv("diabetes_data_clean.csv")

X=df.drop('class', axis =1)
y = df['class']
y.shape, X.shape
```

```python
#split into train and test
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.3, stratify = y, random_state=0)
X_train.shape, y_train.shape
```

```python
# Count the occurrences of each class in the target variable
y_train_counts = y_train.value_counts()
y_test_counts = y_test.value_counts()

print("Training Set Class Distribution:\n", y_train_counts)
print("\nTest Set Class Distribution:\n", y_test_counts)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report, accuracy_score

# Logistic Regression model
logr = LogisticRegression(max_iter=10000)  # Increase iterations to avoid convergence issues
logr.fit(X_train, y_train)

# Predictions
logr_predict = logr.predict(X_test)
logr_predict2 = logr.predict(X_train)

# Confusion Matrix
cm = confusion_matrix(y_test, logr_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

# Accuracy
logr_accuracy_test = accuracy_score(y_test, logr_predict)
print(f"Accuracy of Logistic Regression Model: {logr_accuracy_test:.2f}")
logr_accuracy_train = accuracy_score(y_train, logr_predict2)
print(f"Accuracy of Logistic Regression Model: {logr_accuracy_train:.2f}")
```

```python
import seaborn as sns  # Import seaborn
def plot_classification_report_heatmap(y_true, y_pred, model_name):
    # Generate the classification report
    report = classification_report(y_true, y_pred, output_dict=True)
    
    # Convert the classification report into a DataFrame
    report_df = pd.DataFrame(report).transpose()
    
    # Remove the 'support' column if you don't need it for the heatmap
    report_df.drop('support', axis=1, inplace=True)
    
    # Plot the heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(report_df, annot=True, cmap='coolwarm', fmt='.2f', cbar=False)
    plt.title(f'{model_name} - Classification Report Heatmap')
    plt.show()
```

```python
# Classification Report
print("Classification Report:\n")
print(classification_report(y_test, logr_predict))
plot_classification_report_heatmap(y_test, logr_predict, 'Logistic Regression')
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

```python
# Logistic Regression ROC
logr_probs = logr.predict_proba(X_test)[:, 1]
logr_fpr, logr_tpr, _ = roc_curve(y_test, logr_probs)
logr_auc = auc(logr_fpr, logr_tpr)

plt.figure(figsize=(6, 5))
plt.plot(logr_fpr, logr_tpr, label=f'AUC = {logr_auc:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Logistic Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report, accuracy_score

# Decision Tree model
dt = DecisionTreeClassifier(random_state=0)
dt.fit(X_train, y_train)

# Predictions
dt_predict = dt.predict(X_test)
dt_predict2 = dt.predict(X_train)

# Confusion Matrix
cm = confusion_matrix(y_test, dt_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

# Accuracy
dt_accuracy_test = accuracy_score(y_test, dt_predict)
print(f"Accuracy of Decision Tree Model: {dt_accuracy_test:.2f}")
dt_accuracy_train = accuracy_score(y_train, dt_predict2)
print(f"Accuracy of Decision Tree Model: {dt_accuracy_train:.2f}")
```

```python
# Classification Report
print("Decision Tree Classification Report:\n")
print(classification_report(y_test, dt_predict))
plot_classification_report_heatmap(y_test, dt_predict, 'Decision Tree')
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class (class = 1)
dt_probs = dt.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
dt_fpr, dt_tpr, _ = roc_curve(y_test, dt_probs)
dt_auc = auc(dt_fpr, dt_tpr)

# Plot ROC curve
plt.figure(figsize=(6, 5))
plt.plot(dt_fpr, dt_tpr, label=f'Decision Tree (AUC = {dt_auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # random line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Decision Tree ROC Curve')
plt.legend()
plt.show()

# Print AUC value
print(f"Decision Tree AUC: {dt_auc:.2f}")

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report, accuracy_score

# Random Forest model
rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)

# Predictions
rf_predict = rf.predict(X_test)
rf_predict2 = rf.predict(X_train)

# Confusion Matrix
cm = confusion_matrix(y_test, rf_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

# Accuracy
rf_accuracy_test = accuracy_score(y_test, rf_predict)
print(f"Accuracy of Random Forest Model: {rf_accuracy_test:.2f}")
rf_accuracy_train = accuracy_score(y_train, rf_predict2)
print(f"Accuracy of Random Forest Model: {rf_accuracy_train:.2f}")
```

```python
# Classification Report
print("Random Forest Classification Report:\n")
print(classification_report(y_test, rf_predict))
plot_classification_report_heatmap(y_test, rf_predict, 'Random Forest')
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class (class = 1)
rf_probs = rf.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_probs)
rf_auc = auc(rf_fpr, rf_tpr)

# Plot ROC curve
plt.figure(figsize=(6, 5))
plt.plot(rf_fpr, rf_tpr, label=f'Random Forest (AUC = {rf_auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # random line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest ROC Curve')
plt.legend()
plt.show()

# Print AUC value
print(f"Random Forest AUC: {rf_auc:.2f}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold
import numpy as np

# Create StratifiedKFold to maintain the class ratio in each fold
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

```

```python
# Logistic Regression
logr = LogisticRegression(max_iter=10000)
logr_cv_scores = cross_val_score(logr, X, y, cv=kfold, scoring='accuracy')
print("Logistic Regression - Accuracy per fold:", logr_cv_scores)
print("Logistic Regression - Mean Accuracy:", np.mean(logr_cv_scores))
print("Logistic Regression - Std Deviation:", np.std(logr_cv_scores))
```

```python
# Decision Tree
dt = DecisionTreeClassifier(random_state=0)
dt_cv_scores = cross_val_score(dt, X, y, cv=kfold, scoring='accuracy')
print("Decision Tree - Accuracy per fold:", dt_cv_scores)
print("Decision Tree - Mean Accuracy:", np.mean(dt_cv_scores))
print("Decision Tree - Std Deviation:", np.std(dt_cv_scores))
```

```python
# Random Forest
rf = RandomForestClassifier(random_state=0)
rf_cv_scores = cross_val_score(rf, X, y, cv=kfold, scoring='accuracy')
print("Random Forest - Accuracy per fold:", rf_cv_scores)
print("Random Forest - Mean Accuracy:", np.mean(rf_cv_scores))
print("Random Forest - Std Deviation:", np.std(rf_cv_scores))
```
